```python
import numpy as np
```

```python
class SimpleNeuron:

  def __init__(self, input, weights = None, bias = None):

    if weights is None:
      self.weights = np.random.randn(input) * 0.01
    else:
      self.weights = np.array(weights)


    if bias is None:
      self.bias = 0
    else :
      self.bias = bias


    print(f"  Neuron initialized!")
    print(f"  Weights: {self.weights}")
    print(f"  Bias: {self.bias}")


  def sigmoid(self,z):
    # Sigmoid activation function squashes any value between [0,1] like probability
    return 1/(1+np.exp(-z))


  def process(self,input) :
    z = np.dot(input, self.weights) + self.bias
    # apply activation function

    output = self.sigmoid(z)

    print(f" Output : {output}")
    print(f" Sigmoid : {z}")

    return output, z

  def decide(self, input, threshold = 0.5):
    # Make binary decision based on inputs
    probability, z = self.process(input)
    decision = probability >= threshold

    return decision, probability, z
```

```python
# Spam detection binary classification
```

```python
spam_neuron = SimpleNeuron(input=3)

spam_neuron.weights = np.array([2.0, -1.5, 3.0])  # [has_links, from_known, has_urgency]
spam_neuron.bias = -1.0


# Test emails
test_emails = [
    ("Suspicious email", [0.9, 0.1, 0.8]),  # Many links, unknown sender, urgent
    ("Normal email", [0.2, 0.9, 0.1]),      # Few links, known sender, not urgent
    ("Borderline email", [0.5, 0.5, 0.6])   # Some links, semi-known, somewhat urgent
]


for name, features in test_emails:
    _, prob, z = spam_neuron.decide(features)
    is_spam = "SPAM" if prob > 0.5 else "NOT SPAM"

    print(f"\n{name}:")
    print(f"   Features: {features}")
    print(f"   Linear combination: {z:.3f}")
    print(f"   Spam probability: {prob:.3f} ({prob*100:.1f}%)")
    print(f"   Classification: {is_spam}")
```
